# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).
```

```python
TransformArray = np.array([
    [1, 0, 300],
    [0, 1, 200],
    [0, 0, 1]
])
```

```python
B = []
for point in A:
    new_point = np.dot(TransformArray, np.transpose(point))
    B.append(new_point)
print(B)
```

```python
x = B [:,0]
y = B [:,1]
plt.plot(x, y)
x = A[:,0]
y = A[:,1]
plt.plot(x, y, color='g')
plt.ylim([-200, 400])
plt.xlim([-200, 400])
plt.legend(['новый', 'исходный'])
```
